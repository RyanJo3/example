```python
df_bitcoin = pd.read_csv("BitCoin.csv")
```

```python
df_bitcoin
```

```python
df_bitcoin["5-MA"] = df_bitcoin["Open"].rolling(window=5).mean()
```

```python
df_bitcoin = df_bitcoin[df_bitcoin["Date"].between("2016-06-01", "2017-06-30")]
```

```python
df_bitcoin
```

## Mandatory part
### 2016.6 ~ 2017.6 ê¸°ê°„ì˜ 5-MA ë¹„íŠ¸ì½”ì¸ ê°€ê²© ê·¸ëž˜í”„ ê·¸ë¦¬ê¸°

```python

plt.xlabel("Date")
plt.ylabel("Bitcoin Price")
plt.title("5 - Moving Average Bitcoin Price 2016.06 ~ 2017.06")
sns.lineplot(data=df_bitcoin, x="Date", y="5-MA", color="#f2a900")
plt.show()
```

```python
df_eth = pd.read_csv("ETH_day.csv")
df_eth
```

```python
df_eth["5-MA"] = df_eth["Open"].rolling(window=5).mean()
df_eth = df_eth[df_eth["Date"].between("2016-06-01", "2017-06-30")]
df_eth
```
